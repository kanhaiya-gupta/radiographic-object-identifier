```python
# import required modules

import numpy as np
import matplotlib.pyplot as plt
```

```python
# X and Y values to be plotted
x = [1, 2, 3, 4, 5]          # list of X values
y = [3, 8, 12, 27, 25]       # list of Y values

# Converting the list to array.
# Array can be used for calculation if required

X = np.array(x)     
Y = np.array(y)

# The dimension of X and Y should be same. 
# Only then you can plot. 
print("Length of X: ", len(X))  # Number of points in X
print("Length of Y: ", len(Y))  # Number of points in Y
```

```python
plt.plot(X, Y, label="Data")       # to plot the smooth curve
plt.scatter(x,y)                   # to make the scatter plot
plt.xlabel("X - values")           # Label the X-axis
plt.ylabel("Y - values")           # Label the Y-axis
plt.legend()                       # Show the legend
plt.grid()                         # Show the grid
#plt.savefig("test.png")           # Save the figure
plt.show()                         # Show the plot
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some mock data
t = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(t)
data2 = np.sin(2 * np.pi * t)

temp = data["Unnamed: 0"]
#press = []
#conc = []

fig, ax1 = plt.subplots()

color = 'tab:red'

ax1.plot(t, data1, color=color, linestyle = 'solid', linewidth=2, alpha=1.0)
ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second Axes that shares the same x-axis

color = 'tab:blue'

ax2.plot(t, data2, color=color, linestyle = 'solid', linewidth=2, alpha=1.0)
ax2.set_ylabel('sin', color=color)  # we already handled the x-label with ax1
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
