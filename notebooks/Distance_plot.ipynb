```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle

x_line_1 = [0, 20, 40,  50, 70]
y_line_1 = [0, 0, 0, 0, 0]

x_line_2 = [70, 70, 70,  70, 70, 70]
y_line_2 = [0, 20, 40, 50, 70, 80]

#x_line_3 = [0, 20, 40,  50, 70]
#y_line_3 = [0, 20, 40, 50, 70]

x_line_3 = [0, 70]
y_line_3 = [0, 80]

# Create Figure and Axes
fig,ax = plt.subplots(1)

mid_x, mid_y, max_x, max_y, min_x, min_y,step_size,num_points = 5, 5, 11, 11, 0, 0, 1, 10
x_range = np.concatenate((np.arange(min_x, mid_x, step_size), np.arange(mid_x, max_x, step_size)))
y_range = np.concatenate((np.arange(min_y, mid_y, step_size), np.arange(mid_y, max_y, step_size)))
data = np.array([[x, y] for x in x_range for y in y_range])  # cartesian prod
ax.scatter(data[:, 0]*10, data[:, 1]*10)
ax.scatter(70, 80, color='r')
ax.scatter(0, 0, color='black')
ax.plot(x_line_1, y_line_1, linestyle="dashed")
ax.plot(x_line_2, y_line_2, linestyle="dashed")
ax.plot(x_line_3, y_line_3, linestyle="dashed")

# Add the rectangle
rect = Rectangle( (60,70), 20, 20, linestyle = 'dashed', facecolor = 'None', edgecolor = 'black', clip_on=False)
ax.add_patch(rect)
#ax.grid()


ax.text(36, 2, '70 units', fontsize = 14, weight='normal', rotation=0, rotation_mode='anchor')
ax.text(75, 26, '80 units', fontsize = 14, weight='normal', rotation=90, rotation_mode='anchor')
ax.text(33, 39, '106 units', fontsize = 14, weight='normal', rotation=40, rotation_mode='anchor')
plt.tick_params(axis='x',labelsize=12, colors='black')
plt.tick_params(axis='y',labelsize=12, colors='black')
plt.savefig("../Output/distance_plot.png",  bbox_inches='tight', dpi=400)

plt.show()
```
